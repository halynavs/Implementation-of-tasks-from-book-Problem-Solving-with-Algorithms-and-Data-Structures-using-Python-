### 7. Write a program that can check an HTML document for proper opening and closing tags.

```python
from pythonds.basic import Stack

def check_brakets(input_brakets):
    s=Stack()
    balanced = True
    r=list(input_brakets)
    #print([input_brakets[i:i+1] for i in range(0, len(input_brakets), 1)])
   
    for i in range(len(r)-1):
        if r[i] in '<'  and (not r[i+1] in '/'):
            #print(r[i], 'open',r[i+1])
            s.push(r[i])
        elif r[i] in '/':
            #print(r[i], 'close', r[i+1])
            if s.isEmpty():
                balanced = False
            else:
                top = s.pop()
                if not check_marche_open_close(top, r[i]):
                    balanced = False
        else: 'Your input consists symbols that are not brakets'
            
    if balanced and s.isEmpty():
        return True
    else: 
        return False
    
    
def check_marche_open_close(openb, closeb):
    opens='<'
    closes='/'
    return opens.index(openb)==closes.index(closeb)





print(check_brakets("<html> </html>"))
print(check_brakets("""<html>
   <head>
      <title>
         Example
      </title>
   </head>

   <body>
      <h1>Hello, world</h1>
   </body>
</html>"""))


print(check_brakets("""<html>
   <head>
      <title>
         Example
      </title>
   </head>
      <h2>Good</h2>
   </body>
</html>"""))
```

### 8. Extend the program from previous listing to handle palindromes with spaces. For example, I PREFER PI is a palindrome that reads the same forward and backward if you ignore the blank characters.

```python
from pythonds.basic import Deque

def palindrome_checker(input_string):
    deque_storage=Deque()
    for character in input_string:
        if character !=' ':
            deque_storage.addRear(character)
    
    is_palindrome = True
    
    while is_palindrome and deque_storage.size()>1:       
        if deque_storage.removeRear() != deque_storage.removeFront():
            is_palindrome = False
            
    return is_palindrome

print(palindrome_checker('I PREFER PI'))
print(palindrome_checker('radar'))
print(palindrome_checker('listing to'))
```

```python
# Here Unordered List - collection of nodes


class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
        self.index = 0
        
    def __repr__(self):
        string=''
        string=str(self.data)
        return string
        
    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next
    
    def get_index(self):
        return self.index
    
    def set_data(self, new_data):
        self.data = new_data
        
    def set_next(self, new_next):
        self.next = new_next
        
    def set_index(self, tick):
        self.index += tick
        
class UnorderedList:
    def __init__(self):
        self.head = None
        self.size1 = 0
        self.tail = None
        
    def __repr__(self):
        current = self.head
        string = ''
        while current != None:
            string+=str(current)+' '
            current = current.get_next()
            
        return string
        
    def isEmpty(self):
        return self.head == None
    
    def add(self, item):
        if self.tail == None:
            self.tail = self.head
        temp = Node(item)
        temp.set_next(self.head)
        self.head = temp
        self.size1 +=1
        temp.set_index(self.size1)
        
        
    def length(self):
        return self.size1
        
    def get_tail(self):
        return self.tail
    
    
    def size(self):
        current = self.head
        count = 0
        while current != None:
            count+=1
            current = current.get_next()
            
        return count
    
    def search(self, item):
        is_found = False
        current = self.head
        
        while current != None and (not is_found):
            if current.get_data() == item:
                is_found = True
            else:
                current = current.get_next()
        return is_found
    
    def remove(self, item):
        is_found = False
        current = self.head
        previous = None
        self.size1 -=1
        while (not is_found) and current != None:
            if current.get_data() == item:
                is_found = True
            else:
                previous = current
                current = current.get_next()
                
                
        if previous == None:
            self.head = current.get_next()
        elif current!= None:
            previous.set_next(current.get_next())
            
    def append(self, item):
        if self.tail == None and self.head!=None:
            self.tail = self.head
        
        if self.head == None:
            temp = Node(item)
            temp.set_next(self.head)
            self.head = temp
            temp.set_index(self.size1)
        else:
        
            tail = self.tail
            new =Node(item)
            tail.set_next(new)
            self.tail = new
            
            new.set_index(self.size1)
        
        self.size1 +=1
        
        
    def index(self, item):
        is_found = False
        current = self.head
        
        while current != None and (not is_found):
            if current.get_data() == item:
                is_found = True
            else:
                current = current.get_next()
                
        return current.get_index()
        

        
        
    def pop(self, index=None):
        if index==None or index == -1:   
            index=self.size()
        
        
        is_found = False
        current = self.head
        previous = None
        self.size1 -=1
        while (not is_found) and current.get_next() != None:
            if current.get_index() == index:
                is_found = True
        
            else:
                previous = current
                current = current.get_next()
                
                
        if previous == None:
            self.head = current.get_next()
        elif current!= None:
            previous.set_next(current.get_next())
        
        self.tail = previous
        return current


class Stack:
    
    def __init__(self):
        self.items =UnorderedList()
        
    def __repr__(self):
        return self.items.__repr__()
        
    def isEmpty(self):
        return self.items.isEmpty()
   
    def push(self, new_item):
        self.items.append(new_item)
        
    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items.get_tail()
    
    def size(self):
        return self.items.size()
        
        
myStack = Stack()

myStack.push(100)
myStack.push(11)
myStack.push(345)
print("Instance of class stack named myStack:",myStack)

print("Is myStack Empty?", myStack.isEmpty())

print("Size of myStack:",myStack.size())

print("Removed top from myStack:",myStack.pop())
print("Chenged myStack",myStack)

print("Top of myStack:",myStack.peek())


print("Size of myStack:",myStack.size())
```
